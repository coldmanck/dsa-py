```python
def draw_ruler(num_inches, major_length):
    draw_line(major_length, label='0')
    for i in range(1, num_inches + 1):
        draw_interval(major_length - 1)
        draw_line(major_length, label=str(i))
```

```python
def draw_interval(length):
    if length > 0:
        draw_interval(length - 1)
        draw_line(length)
        draw_interval(length - 1)
```

```python
def draw_line(length, label=''):
    print('-' * length, label)
```

```python
draw_ruler(10, 5)
```
